```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
import numpy as np
```

```python
data = pd.read_csv('D:/study docs/Excel R/Data science/assignment/Logistic regression/Datasets/bank-full.csv')
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
data1=pd.get_dummies(data,columns=['job','marital','education','contact','poutcome','month'])
```

```python
data1
```

```python
data1['default'] = np.where(data1['default'].str.contains("yes"), 1, 0)
data1['housing'] = np.where(data1['housing'].str.contains("yes"), 1, 0)
data1['loan'] = np.where(data1['loan'].str.contains("yes"), 1, 0)
data1['Target'] = np.where(data1['Target'].str.contains("yes"), 1, 0)
```

```python
data1
```

```python
data1.info()
```

```python
x=pd.concat([data1.iloc[:,0:10],data1.iloc[:,11:]],axis=1)
y=data1.Target
```

```python
classifier = LogisticRegression()
classifier.fit(x,y)
```

```python
y_pred = classifier.predict(x)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y,y_pred)
```

```python
confusion_matrix
```

```python
y_pred_df=pd.DataFrame({'actual_y':y,'preicted y':y_pred})
y_pred_df
```

```python
from sklearn.metrics import classification_report
print(classification_report(y,y_pred))
```
